```python
#importing essential libraries
import numpy as np
import pandas as pd
```

```python
#loading my netflix dataset
df = pd.read_csv("NFLX.csv")
df
```

```python
#deleting date and adj close columns
del df["Date"]
del df["Adj Close"]
df
```

```python
#seperating close column
Y = np.array(df["Close"])
Y
```

```python
#deleting close column after seperating in Y variable
del df["Close"]
df
```

```python
#Converting whole dataset into 2-D array
X = df.values
X
```

```python
#importing model_selection from sklearn for spliting X and Y into training and testing
from sklearn import model_selection
X_train, X_test ,Y_train,Y_test = model_selection.train_test_split(X,Y, test_size = 0.3) 
X_train,Y_train
```

```python
print(X_train.shape)
print(Y_train.shape)
```

```python
#importing Linear Regression model from sklearn for predicting closing stockes of netflix
from sklearn.linear_model import LinearRegression
algo = LinearRegression()
algo.fit(X_train,Y_train)
Y_predict = algo.predict(X_test)
print(Y_predict)
print(Y.shape)
```

```python
#printing score of my model
score = algo.score(X_test,Y_test)
score
```
