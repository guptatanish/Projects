```python
movies = movies[['id','title','genres','keywords','overview','cast','crew']]
```

```python
import ast
x = ast.literal_eval(movies.iloc[0].genres)
x
```

```python
def genreconverter(obj):
    x=ast.literal_eval(obj)
    l=[]
    for i in x:
        l.append(i["name"])
    return l
```

```python
movies["genres"]=movies.genres.apply(genreconverter)
```

```python
movies.head(2)
```

```python
movies.iloc[0].keywords
```

```python
def keywordconv(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i["name"])
    return l
```

```python
movies['keywords']=movies.keywords.apply(keywordconv)
```

```python
movies.head(2)
```

```python
movies.iloc[0].cast
```

```python
def castconv(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        l.append(i["name"])
        counter+=1;
        if(counter==5):
            break
    return l
```

```python
movies['cast']=movies.cast.apply(castconv)
```

```python
movies.head(2)
```

```python
movies.iloc[0].crew
```

```python
def crewconv(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=="Director":
            l.append(i["name"])
    return l
```

```python
movies['crew']=movies.crew.apply(crewconv)
```

```python
movies.head(2)
```

```python
def collapse(obj):
    l=[]
    for i in obj:
        l.append(i.replace(" ",""))
    return l
```

```python
movies['genres']=movies.genres.apply(collapse)
movies['keywords']=movies.keywords.apply(collapse)
movies['cast']=movies.cast.apply(collapse)
movies['crew']=movies.crew.apply(collapse)
```

```python
movies.head()
```

```python
movies['overview']=movies.overview.apply(lambda x:x.split())
```

```python
movies['overview']=movies.overview.apply(collapse)
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
new=movies.drop(columns=['keywords','overview','cast','crew'])
```

```python
import nltk
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
import string
stop = stopwords.words('english')
punctuation = list(string.punctuation)
stop = stop + punctuation
```

```python
ps=PorterStemmer()
def stemmed(obj):
    l=[]
    for w in obj:
        if w not in stop:
            l.append(ps.stem(w))
    return l
```

```python
stemmed(movies.iloc[0].tags)
```

```python
new['tags'] = new.tags.apply(stemmed)
```

```python
new.head()
```

```python
new['tags'] = new.tags.apply(lambda x:" ".join(x))
```

```python
new.head()
```

```python
def collapse2(obj):
    return obj.replace(" ","")
```

```python
new["title2"]=new.title.apply(collapse2)
```

```python
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
count_vec = CountVectorizer(max_features=5000)
```

```python
vector = count_vec.fit_transform(new['tags']).toarray()
```

```python
vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity[0:5]
```

```python
similarity.shape
```

```python
new[new['title']=='Spectre'].index[0]
```

```python
x = list(enumerate(similarity[2]))
sorted(x,reverse = True ,key = lambda x:x[1])
```

```python
def recommend(movie,number):
    print("<-----------",number,"Similar Movies are----------->")
    index = new[new['title']==movie].index[0]
    x = list(enumerate(similarity[index]))
    distance = sorted(x,reverse = True ,key = lambda x:x[1])
    c=0
    for i in distance[1:]:
        c+=1
        print(c,"--->",new.iloc[i[0]].title)
        if c==number:
            break
        
```

```python
name = input("Enter the Movies on which you want to Recommend: ")
number = int(input("Enter no of movies you want to see: "))
recommend(name,number)
```

```python
recommend('Harry Potter and the Half-Blood Prince')
```

```python
recommend('The Lord of the Rings: The Fellowship of the Ring')
```
